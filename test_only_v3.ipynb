## **Read CSV file**

<p><strong>Age</strong>: Patients Age in years (Numeric)


<strong>Sex</strong>: Gender (Male : 1; Female : 0) (Nominal)


<strong>cp</strong>: Type of chest pain experienced by patient. This term categorized into 4 category.
0 typical angina, 1 atypical angina, 2 non- anginal pain, 3 asymptomatic (Nominal)

<strong>trestbps</strong>: patient's level of blood pressure at resting mode in mm/HG (Numerical)

<strong>chol</strong>: Serum cholesterol in mg/dl (Numeric)

<strong>fbs</strong>: Blood sugar levels on fasting > 120 mg/dl represents as 1 in case of true and 0 as false (Nominal)


<strong>restecg</strong>: Result of electrocardiogram while at rest are represented in 3 distinct values
0 : Normal 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
2: showing probable or definite left ventricular hypertrophyby Estes' criteria (Nominal)
thalach: Maximum heart rate achieved (Numeric)


<strong>exang</strong>: Angina induced by exercise 0 depicting NO 1 depicting Yes (Nominal)

<strong>oldpeak</strong>: Exercise induced ST-depression in relative with the state of rest (Numeric)
<strong>slope</strong>: ST segment measured in terms of slope during peak exercise
0: up sloping; 1: flat; 2: down sloping(Nominal)
<strong>ca</strong>: The number of major vessels (0–3)(nominal)
<strong>thal</strong>: A blood disorder called thalassemia
0: NULL 1: normal blood flow 2: fixed defect (no blood flow in some part of the heart) 3: reversible defect (a blood flow is observed but it is not normal(nominal)


<strong>target</strong>: It is the target variable which we have to predict 1 means patient is suffering from heart disease and 0 means patient is normal.<p>
